```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y= make_regression(n_samples=5, n_features= 1, n_targets= 1, noise= 80, n_informative=1, random_state= 4)
```

```python
plt.scatter(x,y)
```

```python
#OLS
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x,y)
```

```python
y_pred= lr.predict(x)
```

```python
m=lr.coef_
m
```

```python
b=lr.intercept_
b
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color='red')
```

```python
#Apply gradient descent assuming m=79.98,b=0
y_gd= (79.98*x).reshape(5)
y_pred= lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color='red', label='OLS')
plt.plot(x,y_gd, color='green', label=' b=0')
plt.legend()
plt.show()
```

```python
m=79.98
b=100
slope= -2*np.sum(y- m*x.ravel() -b)
slope
```

```python
lr = 0.01

step_size = slope*lr
step_size
```

```python

# Calculating the new intercept
b = b - step_size
b
```

```python
y_gd1= ((79.98*x)+b).reshape(5)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color='red',label='OLS')
plt.plot(x,y_gd, color='green',label='b=0')
plt.plot(x,y_gd1, color= 'orange',label='b = {}'.format(b))
plt.legend()
plt.show()
```
